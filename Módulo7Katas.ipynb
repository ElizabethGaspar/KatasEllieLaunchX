Comienza agregando dos variables, una para la entrada del usuario, con el nombre `new_planet`, y otra variable para la lista de planetas, denominada `planets`.

```python
# Declara dos variables
new_planet = '' #nuevos planetas
planets = [] #planetas
```

### Crea un ciclo `while`

Comenzando con las variables que acabas de crear, crear谩s un ciclo `while`. El ciclo `while` se ejecutar谩 *mientras* el `new_planet` **no** sea igual a la palabra 'done'.

Dentro del ciclo, comprobar谩s si la variable `new_planet` contiene un valor, que deber铆a ser el nombre de un planeta. Esta es una forma r谩pida de ver si el usuario ha introducido un valor. Si lo han hecho, tu c贸digo agregar谩 (`append`) ese valor a la variable `planets`.

Finalmente, usar谩s `input` para solicitar al usuario que ingrese un nuevo nombre de planeta o que escriba *done* si ha terminado de ingresar nombres de planeta. Almacenar谩 el valor de `input` en la variable `new_planet`.

```python
# Escribe el ciclo while solicitado
while new_planet.lower() != 'done': #.lower minpucula ""recordar"
    if new_planet:
        planets.append(new_planet)
    new_planet = input('Escribe un nuevo planeta ')
```
